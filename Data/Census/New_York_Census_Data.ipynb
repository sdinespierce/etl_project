```python
New_York_census_df_2016=census_df_2016[census_df_2016['STATE'].str.contains("New York")]
New_York_census_df_2016_county = New_York_census_df_2016[["COUNTY_ID","POPULATION","DATE"]]
New_York_census_df_2016_county
```

```python
New_York_census_df_2017=census_df_2017[census_df_2017['STATE'].str.contains("New York")]
New_York_census_df_2017_county = New_York_census_df_2017[["COUNTY_ID","POPULATION","DATE"]]
New_York_census_df_2017_county
```

```python
New_York_census_df_2018=census_df_2018[census_df_2018['STATE'].str.contains("New York")]
New_York_census_df_2018_county = New_York_census_df_2018[["COUNTY_ID","POPULATION","DATE"]]
New_York_census_df_2018_county
```

```python
New_York_census_df_2019=census_df_2019[census_df_2019['STATE'].str.contains("New York")]
New_York_census_df_2019_county = New_York_census_df_2019[["COUNTY_ID","POPULATION","DATE"]]
New_York_census_df_2019_county
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2016_Census_df=New_York_census_df_2016[New_York_census_df_2016['COUNTY_NAME'].str.contains('|'.join(searchfor))]
New_York_City_2016_Census_df = New_York_City_2016_Census_df[["COUNTY_ID","POPULATION","DATE"]]
New_York_City_2016_Census_df
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2017_Census_df=New_York_census_df_2017[New_York_census_df_2017['COUNTY_NAME'].str.contains('|'.join(searchfor))]
New_York_City_2017_Census_df = New_York_City_2017_Census_df[["COUNTY_ID","POPULATION","DATE"]]
New_York_City_2017_Census_df
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2018_Census_df=New_York_census_df_2018[New_York_census_df_2018['COUNTY_NAME'].str.contains('|'.join(searchfor))]
New_York_City_2018_Census_df = New_York_City_2018_Census_df[["COUNTY_ID","POPULATION","DATE"]]
New_York_City_2018_Census_df
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2019_Census_df=New_York_census_df_2019[New_York_census_df_2019['COUNTY_NAME'].str.contains('|'.join(searchfor))]
New_York_City_2019_Census_df = New_York_City_2019_Census_df[["COUNTY_ID","POPULATION","DATE"]]
New_York_City_2019_Census_df
```

```python
Combined_New_York_City_Census_df = pd.concat([New_York_City_2016_Census_df, New_York_City_2017_Census_df, New_York_City_2018_Census_df,New_York_City_2019_Census_df], axis=0)
Combined_New_York_City_Census_df
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
Combined_New_York_City_Census_df.to_csv("ny_census_county_population.csv", encoding="utf-8", index=False)
```
