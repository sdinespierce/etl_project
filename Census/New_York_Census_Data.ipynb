```python
import pandas as pd
```

```python
census_df_2016=pd.read_csv("census_2016_county.csv")
census_df_2016.head()
```

```python
census_df_2017=pd.read_csv("census_2017_county.csv")
census_df_2017.head()
```

```python
census_df_2018=pd.read_csv("census_2018_county.csv")
census_df_2018.head()
```

```python
census_df_2019=pd.read_csv("census_2019_county.csv")
census_df_2019.head()
```

```python
New_York_census_df_2016=census_df_2016[census_df_2016['STATE'].str.contains("New York")]
New_York_census_df_2016.head()
```

```python
New_York_census_df_2017=census_df_2017[census_df_2017['STATE'].str.contains("New York")]
New_York_census_df_2017.head()
```

```python
New_York_census_df_2018=census_df_2018[census_df_2018['STATE'].str.contains("New York")]
New_York_census_df_2018.head()
```

```python
New_York_census_df_2019=census_df_2019[census_df_2019['STATE'].str.contains("New York")]
New_York_census_df_2019.head()
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2016_Census_df=New_York_census_df_2016[New_York_census_df_2016['COUNTY'].str.contains('|'.join(searchfor))]
New_York_City_2016_Census_df
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2017_Census_df=New_York_census_df_2017[New_York_census_df_2017['COUNTY'].str.contains('|'.join(searchfor))]
New_York_City_2017_Census_df
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2018_Census_df=New_York_census_df_2018[New_York_census_df_2018['COUNTY'].str.contains('|'.join(searchfor))]
New_York_City_2018_Census_df
```

```python
searchfor=['New York County','Kings County','Bronx County', 'Richmond County', 'Queens County', 'Suffolk County', 'Nassau County']
New_York_City_2019_Census_df=New_York_census_df_2019[New_York_census_df_2019['COUNTY'].str.contains('|'.join(searchfor))]
New_York_City_2019_Census_df
```

```python
Combined_New_York_City_Census_df = pd.concat([New_York_City_2016_Census_df, New_York_City_2017_Census_df, New_York_City_2018_Census_df,New_York_City_2019_Census_df], axis=0)
Combined_New_York_City_Census_df
```
