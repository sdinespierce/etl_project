```python
# Mike
import pandas as pd
```

```python
#read in race data
income_by_race_df = pd.read_csv('../Data/clean_income_by_location.csv').rename(columns={'County':'COUNTY_NAME'})
income_by_race_df.head()
```

```python
# read in county id
county_id_df = pd.read_csv('../Census/ny_census_county_id.csv')
county_id_df.head()
```

```python
# merge data
household_income_by_race_df =pd.merge(income_by_race_df, county_id_df,on='COUNTY_NAME')
household_income_by_race_df.drop(labels=0, axis=0, inplace=True)
household_income_by_race_df.head()
```

```python
# Race ID Table
income_by_race_df['Race'].unique()
race_id = pd.DataFrame(income_by_race_df['Race'].unique()).rename(columns={0:'Race'})
race_id.drop(labels=0, axis=0, inplace=True)
race_id.index.name='Race ID'
race_id.reset_index(inplace=True)
race_id.head(10)
```

```python
# merge data
merged_race_df =pd.merge(race_id, household_income_by_race_df,on='Race')
merged_race_df.drop(columns='COUNTY_NAME', axis=0, inplace=True)
merged_race_df.head()
```

```python
Household_Income_By_Race = merged_race_df[['Year','Race ID','COUNTY_ID','Household Income by Race', 'Household Income by Race Moe']]
Household_Income_By_Race['Household Income by Race Moe'] = [int(point) for point in Household_Income_By_Race['Household Income by Race Moe']]
Household_Income_By_Race.head()
```

```python
Household_Income_By_Race.to_csv("../Data/db_table_csvs/Household_Income_by_Race.csv", encoding='utf-8',index=False)
```
