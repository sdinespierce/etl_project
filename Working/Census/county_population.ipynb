```python
#T.BRADFORD
#July 2021
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
from datetime import datetime
from census import Census

# Census API Key
from config import api_key
```

```python
# Run Census Search to retrieve population data
url_2019_county ="https://api.census.gov/data/2019/pep/population?get=NAME,POP,DATE_DESC&for=county:*&in=state:*&key="+api_key
response = requests.get(url_2019_county)
response_json = response.json()
```

```python
#Create and rename Dataframe
census_2019_county = pd.DataFrame(response.json()[1:], columns=response.json()[0]).rename(columns=
{"NAME":"COUNTY, STATE", "POP":"POPULATION","DATE_DESC":"DATE","state":"State Code", "county":"COUNTY_ID"})
census_2019_county["POPULATION"].astype(int)

#string split
census_2019_county[['COUNTY_NAME', 'STATE']] = census_2019_county['COUNTY, STATE'].str.split(',', 1, expand=True)
census_2019_county[['DATE', 'Trash']] = census_2019_county['DATE'].str.split(' ', 1, expand=True)

#convert Date to a datetime object
census_2019_county['DATE'] = pd.to_datetime(census_2019_county['DATE'])
#reduced dataframe
census_2019_county = census_2019_county[["COUNTY_ID","COUNTY_NAME","STATE","POPULATION","DATE"]]

census_2019_county
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
census_2019_county.to_csv("census_2019_county.csv", encoding="utf-8", index=False)
```
